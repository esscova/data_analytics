### 4. Você deve ter percebido que as colunas DATA INICIAL e DATA FINAL estão formatadas como `String`.  Utilizando o método `pd.to_datetime()`, converta ambas colunas para Timestamp / Datetime.

### 8. Qual o preço médio de revenda da gasolina em agosto de 2008?

### 12. Você conseguiria obter uma tabela contendo a variação percentual ano a ano para o estado do Rio de Janeiro?

```python
rio=q[(q['DATA FINAL'].dt.month == 12) & (q['ESTADO']=='RIO DE JANEIRO')]
rio.head(3)
```

```python
rio_perc = rio.groupby('ANO-MES')['PREÇO MÉDIO REVENDA'].last().pct_change()*100
rio_perc
```

### DESAFIO: Crie uma tabela contendo uma serie temporal com a diferença absoluta e percentual entre os valores mais baratos e caros. Apresente também ao lado os estados na qual os maiores e menores preços foram registrados.

```python
q.head(3)
```

```python
q_max = q.groupby('ANO-MES').max()['PREÇO MÉDIO REVENDA'].sort_index()
q_min = q.groupby('ANO-MES').min()['PREÇO MÉDIO REVENDA'].sort_index()

q_idx_max = q.groupby('ANO-MES')['PREÇO MÉDIO REVENDA'].idxmax()
q_idx_min = q.groupby('ANO-MES')['PREÇO MÉDIO REVENDA'].idxmin()

df_final = pd.DataFrame()
```

```python
df_final['DIFERENÇA'] = q_max - q_min
df_final['PERCENTUAL'] = (q_max - q_min)/q_min *100
df_final['MAXIMO'] = q_max
df_final['MÍNIMO'] = q_min
```

```python
df_final
```

```python
df_final['ESTADO MAXIMO'] = q.loc[q_idx_max, :]['ESTADO'].values
df_final['ESTADO MINIMO'] = q.loc[q_idx_min, :]['ESTADO'].values
```
