# Visualizando dados com Python - parte III
## Boxplot, setores e barras

```python
# bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### barras

```python
# pandas
df_regiao = pd.DataFrame({
    'Regi達o': ['Norte', 'Sul', 'Leste', 'Oeste'],
    'Produto A': [200, 150, 180, 220],
    'Produto B': [180, 210, 150, 170]
})

df_regiao.set_index('Regi達o').plot(kind='bar', figsize=(10,6), color=['skyblue', 'orange'])
plt.title('Vendas de Produtos por Regi達o')
plt.ylabel('Vendas')
plt.xlabel('Regi達o')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# matplotlib
df = pd.DataFrame(
    {
        'Produto': ['A', 'B', 'C', 'D'],
        'Vendas': [150, 200, 120, 180]
    })
plt.bar(df['Produto'], df['Vendas'], color='skyblue')
plt.title('Vendas por Produto')
plt.xlabel('Produto')
plt.ylabel('Vendas')
plt.show()
```

```python
# seaborn
sns.barplot(x='Produto', y='Vendas', data=df)
plt.title('Vendas por Produto')
plt.xlabel('Produto')
plt.ylabel('Vendas')
plt.show()
```

### setores

```python
# pandas
df.plot.pie(y='Vendas', labels=df['Produto'])
plt.title('Distribui巽達o de Vendas por Produto')
plt.show()
```

```python
# seaborn ate onde procurei n達o tem suporte 測測'
```

```python
# matplotlib
plt.pie(df['Vendas'], 
        labels=df['Produto'],
        autopct='%1.1f%%',
        colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'],
        startangle=90
       )
plt.title('Distribui巽達o de Vendas por Produto')
plt.show()
```

```python
# matplotlib
df_empresas = pd.DataFrame({
    'Empresa': ['A', 'B', 'C', 'D'],
    'Participa巽達o': [35, 25, 20, 20]
})

plt.pie(df_empresas['Participa巽達o'], 
        labels=df_empresas['Empresa'],
        autopct='%1.1f%%',
        colors=['#ff6347','#4682b4','#98fb98','#ffd700'],
        wedgeprops={'edgecolor': 'black', 'linewidth': 2},
        explode=(0.1,0,0,0)
       )

plt.title('Participa巽達o de Mercado das Empresas')
plt.show()
```

### boxplot

```python
# pandas
df_meses = pd.DataFrame({
    'Mes': ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun'],
    'Vendas': [150, 200, 120, 180, 250, 220]
})

df_meses['Vendas'].plot.box()
plt.title('Distribui巽達o das Vendas por M棚s')
plt.show()
```

```python
# seaborn
sns.boxplot(df_meses['Vendas'], color='lightgreen')
plt.title('Boxplot das Vendas')
plt.show()
```

```python
# seaborn
df_idades = pd.DataFrame(
    {'Idade': [23, 45, 34, 25, 67, 89, 32, 28, 39, 48, 50, 54, 35, 60, 40]}
)

sns.boxplot(y=df_idades['Idade'], color='lightblue')
plt.title('Distribui巽達o das Idades dos Clientes')
plt.show()
```
